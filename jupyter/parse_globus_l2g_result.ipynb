```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fn = "globus_log.csv"
```

```python
_df = pd.read_csv(fn, header = None)
```

```python
df = _df.drop(_df.index[12])
```

```python
speeds = list(df[13])
```

```python
speeds = speeds[::-1]
```

```python
for i in range(len(speeds)):
    speeds[i] = float("%.2f" % (speeds[i] / (2**13)))
```

```python
ccList = range(1, 64 + 1)
```

```python
plt.plot(ccList, speeds)
plt.xlabel("cc")
plt.ylabel("GB/s")
ax = plt.axes()        
ax.yaxis.grid()
plt.show()
trMax = max(speeds)
trMaxIdx = speeds.index(trMax) + 1
print("Highest transfer rate achieved is " + str(max(speeds)) + "GB/s, when cc is " + str(trMaxIdx))
```
