```python
file_per_node = [[] for i in range(6)]
for node in range(5, 11):
    fn = "globus_log/dtn" + ("%02d" % node)
    for l in open(fn, "r"):
        lp = l.split()
        #print(lp)
        file_per_node[node-5].append([lp[0], lp[6].split("/")[-1], lp[9]])
```

```python
d = {}
for l in open("globus_log/flist", "r"):
    lp = l.split()
    d[lp[8]] = int(lp[4])
```

```python
size_transfer_per_node = [0] * 6
for i in range(6):
    for fn in file_per_node[i]:
        try:
            size_transfer_per_node[i] += d[fn]
        except:
            pass
```

```python
for i in range(6):
    size_transfer_per_node[i] /= 2**30
```
