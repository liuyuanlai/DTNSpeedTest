```python
import matplotlib.pyplot as plt
import re
```

```python
def new_fig(figsize=(8.9, 5)):
    plt.close()
    fig = plt.figure(figsize=figsize)
    ax1 = fig.add_subplot(111)
    for axis in ['top','bottom','left','right']:
        ax1.spines[axis].set_linewidth(3)    
    ax1.yaxis.grid(linestyle='--')
    for tick in ax1.xaxis.get_major_ticks():
        tick.label.set_fontsize(18) 
    for tick in ax1.yaxis.get_major_ticks():
        tick.label.set_fontsize(18) 
    for line in ax1.get_xticklines() + ax1.get_yticklines():
        line.set_markersize(8)
    return ax1
```

```python
def extract_cwnd(fname, s):
    f = open(fname, 'r')
    lines = f.readlines()
    i = 0
    cwnd = []
    while i < len(lines):
    #print(lines[i])
        if s in lines[i]:
            i += 1
            #print(lines[i])
            m = re.search('cwnd:(\d+)', lines[i], re.IGNORECASE)
            #cwnd.append(int(lines[i].split()[7].split(":")[-1]))
            try:
                cwnd.append(int(m.group(1)))
            except:
                print(lines[i])
        i += 1
    new_fig(figsize = (12, 5))
    plt.plot(range(len(cwnd)), cwnd)
```
